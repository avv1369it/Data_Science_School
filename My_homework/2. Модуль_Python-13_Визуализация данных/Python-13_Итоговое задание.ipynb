````python
# импортируем необходимые библиотеки
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly
import plotly.express as px
````

````python
# импорт данных 
churn_data = pd.read_csv('data/churn.csv')

# выводим предварительную информацию
display (churn_data.info())
````

````python
# выводим информацию по уникальным значениям в столбцах
display(churn_data.nunique())
````

````python
# удаляем столбец с "Номером строки"
churn_data = churn_data.drop('RowNumber', axis=1)

# изменяем тип данных в столбцах: Gender, HasCrCard, NumOfProducts, IsActiveMember на тип данных category
churn_data['Gender'] = churn_data['Gender'].astype('category')
churn_data['HasCrCard'] = churn_data['HasCrCard'].astype('category')
#churn_data['NumOfProducts'] = churn_data['NumOfProducts'].astype('category')
churn_data['IsActiveMember'] = churn_data['IsActiveMember'].astype('category')

# выводим информацию после преобразования для предварительного анализа
display (churn_data.info())
display(churn_data.head(3))
display(churn_data.tail(3))
display(churn_data.describe())
display(churn_data.describe(include=object))
````

````python
# производим корректировку подписи в категориальных данных "IsActiveMember" и "Exited"
churn_data["IsActiveMember"] = churn_data["IsActiveMember"].apply(lambda x: "Неактивные" if x == 0 else "Активные")
churn_data["Exited"] = churn_data["Exited"].apply(lambda x: "Лояльные" if x == 0 else "Нелояльные")
display(churn_data["IsActiveMember"].value_counts())
display(churn_data["Exited"].value_counts())
````

### Задача_10. Переведите числовой признак CreditScore в категориальный. Для этого воспользуйтесь функцией get_credit_score_cat(), которая приведена ниже. Примените её к столбцу CreditScore и создайте новый признак CreditScoreCat — категории кредитного рейтинга.

````python
# функция градации лояльности в зависимости от баллов
def get_credit_score_cat(credit_score):
    if credit_score >= 300 and credit_score < 500:
        return "Very_Poor"
    elif credit_score >= 500 and credit_score < 601:
        return "Poor"
    elif credit_score >= 601 and credit_score < 661:
        return "Fair"
    elif credit_score >= 661 and credit_score < 781:
        return "Good"
    elif credit_score >= 781 and credit_score < 851:
        return "Excellent"
    elif credit_score >= 851:
        return "Top"
    elif credit_score < 300:
        return "Deep"
````

````python
# создание нового признака 
churn_data['Credit_Score_Cat'] = churn_data['CreditScore'].apply(get_credit_score_cat)
display(churn_data['Credit_Score_Cat'].value_counts())
````

````python
# создание таблицы сравнения лояльности клиентов в зависимоти от рейтинга и лет обслуживания
churn_data_10 = churn_data.pivot_table(values='Exited', index='Credit_Score_Cat', columns='Tenure', aggfunc='mean')
display(churn_data_10)
````

````python
# основной график
heatmap = sns.heatmap(data=churn_data_10, cmap='YlOrRd')
heatmap.set_title('Тепловая карта Лояльность & Кредитный рейтинг & Годы оказания услуг ', fontsize=10, color='black')
````
